```python
df_spark=spark.sql("""
WITH latest_date AS (
  SELECT
    ETF,
    MAX(Date) AS max_date
  FROM my_stock_data.my_stock_schema.Source_histry_etf
  GROUP BY ETF
),
periods AS (
  SELECT
    ETF,
    max_date,
    DATE_SUB(max_date, 365) AS year_ago,
    DATE_SUB(max_date, 180) AS half_year_ago,
    DATE_SUB(max_date, 90) AS quarter_ago,
    DATE_SUB(max_date, 30) AS month_ago,
    DATE_SUB(max_date, 7) AS week_ago
  FROM latest_date
),
prices AS (
  SELECT
    s.ETF,
    s.Date,
    s.Close,
    p.max_date,
    p.year_ago,
    p.half_year_ago,
    p.quarter_ago,
    p.month_ago,
    p.week_ago
  FROM my_stock_data.my_stock_schema.Source_histry_etf s
  JOIN periods p ON s.ETF = p.ETF
  WHERE s.Date = p.max_date
     OR s.Date BETWEEN p.year_ago AND DATE_ADD(p.year_ago, 5)
     OR s.Date BETWEEN p.half_year_ago AND DATE_ADD(p.half_year_ago, 5)
     OR s.Date BETWEEN p.quarter_ago AND DATE_ADD(p.quarter_ago, 5)
     OR s.Date BETWEEN p.month_ago AND DATE_ADD(p.month_ago, 5)
     OR s.Date BETWEEN p.week_ago AND DATE_ADD(p.week_ago, 5)
),
aggregated AS (
  SELECT
    ETF,
    MAX(CASE WHEN Date = max_date THEN Close END) AS latest_close,
    MAX(CASE WHEN Date BETWEEN year_ago AND DATE_ADD(year_ago, 5) THEN Close END) AS year_ago_close,
    MAX(CASE WHEN Date BETWEEN half_year_ago AND DATE_ADD(half_year_ago, 5) THEN Close END) AS half_year_ago_close,
    MAX(CASE WHEN Date BETWEEN quarter_ago AND DATE_ADD(quarter_ago, 5) THEN Close END) AS quarter_ago_close,
    MAX(CASE WHEN Date BETWEEN month_ago AND DATE_ADD(month_ago, 5) THEN Close END) AS month_ago_close,
    MAX(CASE WHEN Date BETWEEN week_ago AND DATE_ADD(week_ago, 5) THEN Close END) AS week_ago_close,
    MAX(max_date) AS latest_date
  FROM prices
  GROUP BY ETF
)
SELECT
  ETF,
  latest_date,
  latest_close,
  year_ago_close,
  half_year_ago_close,
  quarter_ago_close,
  month_ago_close,
  week_ago_close,
  ROUND((latest_close - year_ago_close) / year_ago_close, 4) AS yearly_change,
  ROUND((latest_close - half_year_ago_close) / half_year_ago_close, 4) AS half_yearly_change,
  ROUND((latest_close - quarter_ago_close) / quarter_ago_close, 4) AS quarterly_change,
  ROUND((latest_close - month_ago_close) / month_ago_close, 4) AS monthly_change,
  ROUND((latest_close - week_ago_close) / week_ago_close, 4) AS weekly_change
FROM aggregated""")
```

```python
df_spark.write.mode("overwrite").saveAsTable("my_stock_data.my_stock_schema.Source_histry_etf_analysis")
```
